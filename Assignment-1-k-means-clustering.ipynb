```python
#import necessary libraries
import pandas as pd 
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt 
%matplotlib inline

dataset=[[21,65], [65,40],[5,75],
         [25,70], [60,70],[15,70],
         [10,55], [60,50],[40,10],
         [35,5],  [48,10], [20,75],
         [40,20], [25,79],[34,51],
         [22,53], [27,78],[33,59],
         [33,74], [31,73], [22,57],
         [35,69], [34,75],[67,51],
         [54,32], [57,40],[43,47],
         [50,53], [57,36],[59,35],
         [52,58], [65,59],[47,50],
         [49,25], [48,20],[35,14],
         [33,12], [44,20],[45,5],
         [38,29], [43,27],[51,8],
         [46,7], [7,65],  [50,40],
         [15,60], [5,50], [30,30]
]

#Create feature vector X by combining two columns into single one matrix
X= np.array(dataset)

plt.scatter(X[:,0], X[:,1], s=50, c='red')
plt.title("Before Applying K-Means")

plt.show()  
# The KMeans algorithm clusters data by trying to separate samples in n groups of equal variance.
#Apply K-Means algorithm to identify clusters (here n is number of clusters selected i.e. 2,3,4) n=2
kmeans = KMeans(n_clusters=2).fit(X)                
centroids = kmeans.cluster_centers_

       #Plot of data points after KMeans non clustered has been applied    
#  (NOTE: Colors here give cluster identification on plot )               
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_.astype(float),s=50,alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.title("Clusters with k=2")
plt.show()
print("Centroids:",centroids)
#Apply K-Means algorithm to identify clusters (here n is number of clusters selected i.e. 2,3,4) n=3
kmeans = KMeans(n_clusters=3).fit(X)                
centroids = kmeans.cluster_centers_

plt.scatter(X[:,0], X[:,1], c=kmeans.labels_.astype(float),s=50,alpha=0.5) 
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.title("Clusters with k=3")
plt.show()
print("Centroids:",centroids)

#Apply K-Means algorithm to identify clusters (here n is number of clusters selected i.e. 2,3,4) n=4
kmeans = KMeans(n_clusters=4).fit(X)                
centroids = kmeans.cluster_centers_

plt.scatter(X[:,0], X[:,1], c=kmeans.labels_.astype(float),s=50,alpha=0.5) 
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.title("Clusters with k=4")
plt.show()
print("Centroids:",centroids)
```
